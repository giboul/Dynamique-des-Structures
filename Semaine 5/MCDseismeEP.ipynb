```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('bmh')
```

```python
## Calcul de la reponse d'un systeme lineaire
## par la methode de la difference centrale
## (Chopra p.161)
## 
## Accelerogramme dans Seisme
## 
ulin=44.285e-3
R=3.0
## interval de temps: dt
dt=1/100
## 
## nb de points: nb
nb=4000
t = np.arange(0,nb)*dt
Seisme = np.zeros(nb)
xmax = 10*np.pi
x = np.linspace(0,xmax,nb)
amplitude = 10
Seisme = np.sin(x)*(xmax - x) / xmax *amplitude
## 
## amortissement: z
z=5.0/100
## 
## frequence propre: f
f=2.0
## Stabilite
'Tn/dt (> pi)'
1/f/dt
if 1/f/dt <= np.pi:
    raise ValueError(f'{1/f/dt = } <= pi')
## pulsation propre: w
w=2*np.pi*f
## 
## conditions initiales
u = np.zeros(nb)
v = np.zeros(nb)
## 
## 
## ****************************************************************
## Calcul initiaux kksurM, asurM, bsurM, uzero
## ****************************************************************
## 
## calculs preliminaires
a = np.zeros(nb)
k = np.zeros(nb)
a[0]=Seisme[0]-2*z*w*v[0]-w**2*u[0]
uzero=u[0]-dt*v[0]+dt**2/2*a[0]
kksurM=1/dt**2+z*w/dt
asurM=1/dt**2-z*w/dt
## bsurM=w**2-2/dt**2
## 
## ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
bsurM=-2/dt**2
## ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
## 
## 
## ****************************************************************
## deplacements u, vitesses v, accelerations a
## ****************************************************************
## 
## ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
pivot=0
yield_ = np.zeros(nb)
mud = np.zeros(nb)
## ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
## 
I=1
ppsurM=Seisme[I]-asurM*uzero-bsurM*u[I]
u[I+1]=ppsurM/kksurM
v[I]=u[I+1]-uzero/2/dt
a[I]=(u[I+1]-2*u[I]+uzero)/dt**2

for I in range(1,nb-1):
    
    if abs(u[I]-pivot) > ulin/R:
        kmod=np.sign(u[I]-pivot)*w**2/R*ulin
        pivot=u[I]-np.sign(u[I]-pivot)*ulin/R
        yield_[I]=np.sign(u[I]-pivot) 
    else:
        kmod=w**2*(u[I]-pivot)

    k[I]=kmod
    ppsurM=Seisme[I]-asurM*u[I-1]-bsurM*u[I]-kmod
    ## ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
    ## 
    ## 	ppsurM=Seisme[I]-asurM*u[I-1]-bsurM*u[I]
    u[I+1]=ppsurM/kksurM
    v[I]=(u[I+1]-u[I-1])/2/dt
    a[I]=(u[I+1]-2*u[I]+u[I-1])/dt**2

v[nb-1]=v[nb-2]
a[nb-1]=a[nb-2]
## 
## dessin de la reponse
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(t,u,'r', linewidth=0.8)
ax1.set_ylabel('$u$')
ax1.yaxis.label.set_color('r')
ax1.tick_params(colors='r', which='both')

ax2.plot(t, Seisme, '--b', linewidth=0.8)
ax2.set_ylabel('Seisme')
ax2.yaxis.label.set_color('b')
ax2.tick_params(colors='b', which='both')

plt.show()
```
